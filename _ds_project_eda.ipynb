## Data By Artist Visualization and Exploration

```python
df_byartist
```

```python
# Exploring range of the numerical features
fig, axs = plt.subplots(3, 3, figsize=(18, 15))
features = ['popularity', 'danceability', 'energy', 'valence', 'tempo', 'acousticness', 'instrumentalness', 'speechiness', 'loudness']
for i, feature in enumerate(features):
    sns.histplot(df_byartist[feature], kde=True, ax=axs[i//3, i%3], color="skyblue")
    axs[i//3, i%3].set_title(f'Distribution of {feature}')

plt.tight_layout()
plt.show()
```

```python
# Exploring correlation among the numerical features
correlation_matrix = df_byartist[features].corr()

plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", vmin=-1, vmax=1, fmt=".2f")
plt.title("Correlation Matrix of Numerical Features")
plt.show()
```

### Inferences:

1. **Energy vs Loudness:** Strong positive correlation as expected, since in general energetic music tends to be louder.
2. **Valence vs Danceability:** Medium to strong positive correlation indicating that danceable songs are more positive in tone.
3. **Acousticness vs Energy, Danceability, Popularity, Loudness:** Strong negative correlation, indicating that acoustic music will not be energetic and loud or danceable and hence less popular as well.
4. **Popularity vs Other Features:** Very low negative and positve correlation, indicating that popularity is not highly dependent on the numerical features.

```python
# Popularity based on Mode
plt.figure(figsize=(8, 6))
sns.boxplot(x='mode', y='popularity', data=df_byartist, palette="pastel")
plt.title("Popularity based on Mode (Major or Minor)")
plt.xlabel("Mode")
plt.ylabel("Popularity")
plt.show()
```

### Inference

Popularity has a higher overall range for Mode 0 than Mode 1

```python
popularity_by_artist = df_byartist.groupby('artists')['popularity'].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(7, 4))
sns.barplot(x=popularity_by_artist.values, y=popularity_by_artist.index, color="lightpink")
plt.title("Top 10 Individual Artists by Average Popularity")
plt.xlabel("Average Popularity")
plt.ylabel("Genre")
plt.show()
```

## Data By Genres Visualization and Exploration

```python

sns.set(style="whitegrid")

plt.figure(figsize=(7, 4))
sns.scatterplot(x='danceability', y='popularity', data=df_bygenres, color='purple', alpha=0.6)
plt.title('Danceability vs Popularity')
plt.xlabel('Danceability')
plt.ylabel('Popularity')
plt.show()

plt.figure(figsize=(7, 4))
sns.scatterplot(x='energy', y='popularity', data=df_bygenres, color='coral', alpha=0.6)
plt.title('Energy vs Popularity')
plt.xlabel('Energy')
plt.ylabel('Popularity')
plt.show()
```

```python
popularity_by_genre = df_bygenres.groupby('genres')['popularity'].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(7, 4))
sns.barplot(x=popularity_by_genre.values, y=popularity_by_genre.index, color="skyblue")
plt.title("Top 10 Genres by Average Popularity")
plt.xlabel("Average Popularity")
plt.ylabel("Genre")
plt.show()
```

```python
genre_grouped = df_bygenres.groupby('genres').mean().reset_index()

top_10_genres = genre_grouped.nlargest(10, 'popularity')

plt.figure(figsize=(16, 6))
plt.bar(top_10_genres['genres'], top_10_genres['acousticness'])
plt.title('Average Acousticness by Genre (Top 10)')
plt.xlabel('Genre')
plt.ylabel('Acousticness')
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()

plt.figure(figsize=(16, 6))
plt.bar(top_10_genres['genres'], top_10_genres['danceability'])
plt.title('Average Danceability by Genre (Top 10)')
plt.xlabel('Genre')
plt.ylabel('Danceability')
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()

plt.figure(figsize=(16, 6))
plt.bar(top_10_genres['genres'], top_10_genres['energy'])
plt.title('Average Energy by Genre (Top 10)')
plt.xlabel('Genre')
plt.ylabel('Energy')
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()

plt.figure(figsize=(16, 6))
plt.bar(top_10_genres['genres'], top_10_genres['valence'])
plt.title('Average Valence by Genre (Top 10)')
plt.xlabel('Genre')
plt.ylabel('Valence')
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()

plt.figure(figsize=(16, 6))
plt.bar(top_10_genres['genres'], top_10_genres['duration_ms'])
plt.title('Average Duration by Genre (Top 10)')
plt.xlabel('Genre')
plt.ylabel('Duration (ms)')
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()

plt.figure(figsize=(16, 6))
plt.bar(top_10_genres['genres'], top_10_genres['popularity'])
plt.title('Popularity Distribution by Genre (Top 10)')
plt.xlabel('Genre')
plt.ylabel('Popularity')
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

**Inference:**

The popularity vs genre graph shows that genres like basshall, trap veenezolano, south african house and turkish edm are among the most popular. These genres have high values of valence, energy, and danceability.

High valence in these genres indicates a positive and happy mood. These genres can be recommended to the users looking for uplifting and cheerful music.

From the graphs it can also be inferred that the tops genres have high energy and danceability. This makes the genres preferred by users seeking tracks that are energetic, rhythmic and have high beats.

It can also be observed that acousticness graph is lower for these popular genres. This suits users who prefer modern, electronic, or digitally produced music.

Using these insights, we can match users to genres based on their preferences for energy, danceability, and mood.
