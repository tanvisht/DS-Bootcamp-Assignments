4. How do you control for biases?
Controlling biases involves recognizing and mitigating factors that may distort results:

Randomization: Assign subjects or samples randomly to groups or treatments.
Blinding: Use single or double-blind methods to prevent expectations from influencing outcomes.
Standardization: Ensure consistent data collection and measurement procedures.
Control Groups: Include groups that do not receive treatment to serve as a baseline.
Data Cleaning: Identify and remove outliers or errors.
Awareness of Cognitive Biases: Be vigilant against confirmation bias, selection bias, or survivorship bias during analysis.
5. What are confounding variables?
Confounding variables are extraneous factors that can distort the apparent relationship between an independent variable and a dependent variable.

Example: In a study linking exercise (independent variable) to weight loss (dependent variable), diet could be a confounding variable if not controlled.
Effects: They can lead to false associations or obscure true relationships.
Mitigation:
Randomization in experimental design.
Statistical techniques like stratification or multivariate regression.
Matching groups based on confounder characteristics.
6. What is A/B testing?
A/B testing is a statistical method used to compare two versions (A and B) of a variable (e.g., website design, marketing email) to determine which performs better.

Steps:
Define a goal (e.g., click-through rate).
Split participants randomly into two groups.
Expose each group to version A or B.
Collect data on the outcome metric.
Analyze results to determine if differences are statistically significant.
Use Cases: Widely used in product development, marketing, and UX design to optimize decisions based on empirical evidence.
